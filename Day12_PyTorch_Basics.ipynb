```python
import torch
import torch.nn as nn
class Mediscannet(nn.Module):
   def __init__(self):
       super().__init__()
       self.layer1=nn.Linear(3,64)
       self.layer2=nn.Linear(64,32)
       self.layer3=nn.Linear(32,3)
       self.relu=nn.ReLU()
   def forward(self,x):
    x=self.relu(self.layer1(x))
    x=self.relu(self.layer2(x))
    x=self.layer3(x)
    return x

model=Mediscannet()
print(model)

```
