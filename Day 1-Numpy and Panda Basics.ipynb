```python
import pandas as pd
data={
    'medicine' : ['paracetamol','panadol','aspirin'],
    'riskscore' :[2,5,8],
    'risklevel' :['low','medium','high']
}
df=pd.DataFrame(data)
high_risk=df[df['risklevel']=='high']
print(high_risk)

risky = df[df['riskscore'] > 5]
print(risky)
```
