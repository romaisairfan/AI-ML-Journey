```python
import torch
import torch.nn as nn

image=torch.randn(1,3,224,224)
print(f"Image shape:{image.shape}")
print(f"Total pixels:{3*224*224}")
conv_layer=nn.Conv2d(
    in_channels=3,
    out_channels=16,
    kernel_size=3,
    padding=1
)
output=conv_layer(image)
print(f"image shape:{image.shape}")
print(f"output shape:{output.shape}")
pool_layer=nn.MaxPool2d(kernel_size=2,stride=2)
pooled=pool_layer(output)
print(f"after convo output shape:{output.shape}")
print(f"after pooling output shape:{pooled.shape}")
```

```python
conv_layer=nn.Conv2d(
    in_channels=3,
    out_channels=16,
    kernel_size=3,
    padding=1
)
class MediScanCNN(nn.Module):
  def __init__(self):
    super().__init__()
    self.convo1=nn.Conv2d(3,16,kernel_size=3,padding=1)
    self.convo2=nn.Conv2d(16,32,kernel_size=3,padding=1)

    self.pool=nn.MaxPool2d(2,2)
    self.relu=nn.ReLU()
    self.fc1=nn.Linear(32*56*56,128)
    self.fc2=nn.Linear(128,3)

  def forward(self,x):
    x=self.pool(self.relu(self.convo1(x)))
    x=self.pool(self.relu(self.convo2(x)))
    x = x.view(-1, 32 * 56 * 56)
    x=self.relu(self.fc1(x))
    x=self.fc2(x)
    return x
model=MediScanCNN()
print(model)
```
