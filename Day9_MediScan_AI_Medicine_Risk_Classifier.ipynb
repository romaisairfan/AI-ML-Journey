```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report
np.random.seed(42)
n=500
data={
    'dosage_mg':np.random.randint(50,1000,n),
    'interactions':np.random.randint(0,10,n),
    'side_effects_score':np.random.uniform(0,10,n).round(1),
    'age_group':np.random.choice(['child','adult','elderly'],n),
    'chronic_condition':np.random.choice(['yes','no'],n),
}
df=pd.DataFrame(data)
print(df.head())
print(df.shape)
def assign_risk(row):
  score=0
  if row['dosage_mg']>500:
       score+=2
  if row['interactions']>5:
    score+=2
  if row['side_effects_score']>7:
    score+=2
  if row['age_group']=='child' or row['age_group']=='elderly':
     score+=1
  if row['chronic_condition']=='yes':
    score+=1


  if score>=5:
    return'High'
  elif score>=3:
    return'Medium'
  else:
    return 'Low'
df['risk_level']=df.apply(assign_risk,axis=1)
print(df['risk_level'].value_counts())
print(df.head())
le=LabelEncoder()
df['age_group']=le.fit_transform(df['age_group'])
df['chronic_condition']=le.fit_transform(df['chronic_condition'])
y=df['risk_level']
x=df.drop('risk_level',axis=1)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=RandomForestClassifier()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print(f"model accuracy:{accuracy*100:.2f}%")
print("Classification Report")
print(classification_report(y_test,y_pred))


```
