```python
# Day 10 — Revision & Interview Prep

## Phase 1 & 2 Complete Review

### Q1: What is EDA?
Exploratory Data Analysis — we explore, analyze and visualize data before training any ML model to understand patterns, missing values and data distribution.

### Q2: How do you handle missing values?
- Large number of missing values → Drop the column (axis=1)
- Small number of missing values → Fill with mean (numbers) or mode (strings) using fillna()

### Q3: Why is Label Encoding important?
ML models only understand numbers, not text. Label Encoding converts text to numbers e.g. male→1, female→0, Low→0, Medium→1, High→2.

### Q4: What is Train/Test Split?
We divide data 80/20 — 80% for training (model learns) and 20% for testing (model is evaluated on unseen data).

### Q5: What is Overfitting?
When model cramps training data but fails on new data. Sign: huge gap between training accuracy (98%) and testing accuracy (78%). Fix: max_depth parameter.

### Q6: What is Random Forest?
Ensemble of hundreds of Decision Trees — like hundreds of doctors treating one patient and voting for final diagnosis. More trees = better accuracy!

### Q7: What is KNN?
K Nearest Neighbors — model finds K most similar data points and predicts based on majority. Requires Feature Scaling!

### Q8: Recall vs Precision — MediScan AI?
Recall is more important — if model predicts dangerous medicine as safe, patient faces health risk. False Negatives are dangerous!

### Q9: What is Hyperparameter Tuning?
Finding the best model settings e.g. best K value in KNN. K=7 gave best accuracy (82%) compared to K=3,5,9,11.

### Q10: Model A vs Model B?
- Model A: 95% train, 70% test → Overfitting! ❌
- Model B: 82% train, 80% test → Better! Small gap = good generalization ✅

## MediScan AI Project Description
MediScan AI is a medicine risk screening PWA built with React frontend and Python backend. It works in 3 layers:
1. Barcode scanning using ZXing — checks medicine against registry
2. Image analysis using EfficientNet-B0 CNN — analyzes packaging
3. OCR + AI text analysis — extracts text from packaging

All three layers combine scores using weighted fusion to give final risk verdict: No Anomalies, Suspicious, or High Risk.

ML Component: Random Forest classifier that classifies medicines as Low/Medium/High risk based on dosage, interactions, side effects, age group and chronic conditions.
```
