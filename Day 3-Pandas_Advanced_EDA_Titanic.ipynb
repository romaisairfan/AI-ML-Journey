```python
import pandas as pd
table1=pd.DataFrame({
    'medicine':['paracetamol','panadol'],
    'risk_score':[2,5]
})
table2=pd.DataFrame({
    'medicine':['aspirin','gaviscon'],
    'risk_score':[8,3]
})
result=pd.concat([table1,table2],ignore_index=True)
print(result)

```

```python
import pandas as pd
medicine_info=pd.DataFrame({
    'Medicine_ID':[1,2,3],
    'Medicine':['paracetamol','panadol','bruphen']
})
risk_table=pd.DataFrame({
    'Medicine_ID':[1,2,3],
    'risk_level':['low','medium','high']
})
result=pd.merge(medicine_info, risk_table , on='Medicine_ID')
print(result)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df=pd.read_csv(url)
print(df.shape)
print(df.info())
print(df.isnull().sum())
```

```python
import pandas as pd
df=df.drop('Cabin', axis=1)
df['Age']=df['Age'].fillna(df['Age'].mean())
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
print(df.isnull().sum())
```

```python
import pandas as pd
print(df['Survived'].value_counts())
print(df.groupby('Sex')['Age'].mean())
print(df.groupby('Pclass')['Survived'].mean())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
fig,axes=plt.subplots(1,3,figsize=(15,5))
df['Survived'].value_counts().plot(kind='bar',ax=axes[0])
axes[0].set_title('Survival Count')
sns.countplot(x='Survived', hue='Sex',data=df,ax=axes[1])
axes[1].set_title('Survival by Gender')
sns.countplot(x='Survived',hue='Pclass',data=df,ax=axes[2])
axes[2].set_title('Survival by Class')
plt.tight_layout()
plt.show()
```
