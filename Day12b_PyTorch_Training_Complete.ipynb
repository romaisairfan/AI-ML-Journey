```python
import torch
import torch.nn as nn
criterian = nn.CrossEntropyLoss()
prediction=torch.tensor([[0.2,0.7,0.1]])
actual=torch.tensor([1])
loss=criterian(prediction,actual)
print(f"Loss:{loss.item():.2f}")
```

```python
import torch.optim as optim
class SimpleNet(nn.Module):
      def __init__(self):
        super().__init__()
        self.layer1=nn.Linear(3,64)
        self.layer2=nn.Linear(64,3)
        self.relu=nn.ReLU()
      def forward(self,x):
        x=self.relu(self.layer1(x))
        x=self.layer2(x)
        return x
model=SimpleNet()
optimizer=optim.Adam(model.parameters(),lr=0.001)
print("model and optimizer is ready")

```

```python
import numpy as np
X = torch.tensor([[152, 9, 8.5],
                  [485, 6, 0.9],
                  [910, 8, 8.1],
                  [320, 3, 0.6]], dtype=torch.float32)
y=torch.tensor([2,1,2,0])
criterion=nn.CrossEntropyLoss()
for epoch in range(100):
  output=model(X)
  loss=criterion(output,y)
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
  if epoch%10==0:
    print(f"epoch{epoch}:loss={loss.item():4f}")
```

```python
new_medicine=torch.tensor([[500,7,8.0]], dtype=torch.float32)
with torch.no_grad():
  output=model(new_medicine)
  print(output)
  prediction=output.argmax()
  print(prediction)
classes=['low','medium','high']
print(f"risk_level:{classes[prediction]}")
```
